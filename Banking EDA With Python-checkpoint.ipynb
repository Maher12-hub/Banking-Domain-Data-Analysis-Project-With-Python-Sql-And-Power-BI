```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('Banking.csv')
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Estimated Income'].min()
```

```python
#From Numerical to Categorical
bins=[0,100000,300000, float('inf')]
labels=['Low', 'Med', 'High']
df['Income Band']= pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
#Examine the distribution of unique categories in categorical columns
categorical_cols= df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned", "Risk Weighting", "Income Band"]]
for col in categorical_cols:
    print(f"value counts for '{col}':")
    display(df[col].value_counts())
```

```python
#univariate analysis
for i, predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned", "Risk Weighting", "Income Band"]]):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
```

```python
#univariate analysis
for i, predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned", "Risk Weighting", "Income Band"]]):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue="GenderId")
```

```python
#univariate analysis
for i, predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned", "Risk Weighting", "Income Band"]]):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue="Nationality")
```

```python
#HIstplot of value counts for different columnsf
for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title(f'Histogram of {col} Count')
    plt.xlabel(col)
    plt.ylabel("count")
    plt.show()
```

```python
#Numerical Analysis of the datasets where the column values are numerical
numerical_cols=['Estimated Income', 'Superannuation Savings','Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']
plt.figure(figsize=(25,20))
for i,col in enumerate(numerical_cols):
    plt.subplot(3,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()
```

```python
#Heatmaps- correlation between numerical columns
numerical_cols=['Estimated Income', 'Superannuation Savings','Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']
correlation_matrix= df[numerical_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt=".2f")
plt.title("correlation Matrix")
```

#insights of EDA
The strongest positive correlation occur among "Bank Deposits" with "Checking Accounts", "Savings Accounts" and "Foreign Currency Account" indicating that customers who maintain high balances in one account type often held substantial amount/funds across other accounts as well.
